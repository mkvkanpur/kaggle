# 2D Python code (loop)

```python
import numpy as np
import matplotlib.pyplot as plt
        
def compute_spectrum_loop(N, uk, ek):
    for i in range(N[0]):
        for ky in range(1,N[1]//2+1):  # ky > 0
            kx = i if i < N[0]//2+1 else i-N[0]
            k_ind = int(np.ceil(np.sqrt(kx**2+ky**2)))
            ek[k_ind] += (np.abs(uk[i,ky]))**2

    for i in range(N[0]):  # ky =0
        kx = i if i < N[1]//2+1 else i-N[1]
        ek[abs(kx)] += (np.abs(uk[i,0]))**2/2


def set_vel_field(N, dx, dy, ux):
    for i in range(N[0]):
        for j in range(N[1]):
            x = i*dx; y = j*dy
            ux[i,j] = 2+ 2*np.sin(y) + 4*np.sin(x)*np.cos(y)

N = (8,8)
kmax = int(np.ceil(np.linalg.norm(N))//2)+1
ek = np.zeros(kmax)

dx = 2*np.pi/N[0]
dy = 2*np.pi/N[1]
ux = np.zeros(N)
set_vel_field(N, dx, dy, ux)

ukx = np.fft.rfft2(ux)/(N[0]*N[1]) 
#print((ukx))
compute_spectrum_loop(N, ukx, ek)
print("ek ",  ek)

# plt.figure()
# for q in range(qmin,qmax+1):
#     plt.loglog(xforsf, sf[q-qmin,:], label='$d = %i$' %q)

# plt.legend(loc = 'upper left')

```

# 2D Numba CUDA (Global Mem)

```python
Mydatatype = "float64"
from numba import cuda, float64, int32
import numpy as np
import matplotlib.pyplot as plt
import math
import h5py as hp
import time

TPB = (2,2)
##CUDA kernel
@cuda.jit
def compute_spectrum(N, uk, ek):

    (i,ky) = cuda.grid(2)
    kx = i if i < N[0]//2+1 else i-N[0]
    k_ind = int(math.ceil(math.sqrt(kx**2+ky**2)))
    cuda.syncthreads()
    modal_ek = uk[i,ky].real**2 + uk[i,ky].imag**2
    if ky==0:
        modal_ek /= 2
    cuda.atomic.add(ek, k_ind, modal_ek)


### def vel field
def set_vel_field(N, dx, dy, ux):
    for i in range(N[0]):
        for j in range(N[1]):
            x = i*dx; y = j*dy
            ux[i,j] = 2+ 2*np.sin(y) + 4*np.sin(x)*np.cos(y)

### MAIN
N = (8,8)
kmax = int(np.ceil(np.linalg.norm(N))//2)+1
ek = np.zeros(kmax)

dx = 2*np.pi/N[0]
dy = 2*np.pi/N[1]
ux = np.zeros(N)
set_vel_field(N, dx, dy, ux)

ukx = np.fft.rfft2(ux)/(N[0]*N[1]) 

import math
BPG_x = math.ceil(N[0]/(TPB[0]))
BPG_y = math.ceil(N[1]/(TPB[1]))
BPG = (BPG_x, BPG_y)
print(BPG)
#Copy the arrays to the device
ukx_global_mem = cuda.to_device(ukx)
ek_global_mem = cuda.to_device(ek)

#start the kernel
compute_spectrum[BPG,TPB](N, ukx_global_mem, ek_global_mem)
cuda.synchronize()

# Copy the result back to the host
ek = ek_global_mem.copy_to_host()
print(ek)
```

# 3D Numba CUDA (GLOBAL MEM)
Note: np.fft.rfftn is not supported in numba cuda. So we are using cp.fft.rfftn

```python
Mydatatype = "float64"
from numba import cuda, float64, int32
import numpy as np
import matplotlib.pyplot as plt
import math
import h5py as hp
import time
import cupy as cp

TPB = (4,4,4)
##CUDA kernel
@cuda.jit
def compute_spectrum(N, uk, ek):

    (i, j, kz) = cuda.grid(3)
    kx = i if i < N[0]//2+1 else i-N[0]
    ky = j if j < N[1]//2+1 else j-N[1]
    k_ind = int(math.ceil(math.sqrt(kx**2+ ky**2+ kz**2)))
    cuda.syncthreads()
    
    modal_ek = uk[i,j,kz].real**2 + uk[i,j,kz].imag**2
    if kz==0:
        modal_ek /= 2
    cuda.atomic.add(ek, k_ind, modal_ek)

def set_vel_field(N, dx, dy, dz, ux):
    for i in range(N[0]):
        for j in range(N[1]):
            for k in range(N[2]):
                x = i*dx; y = j*dy; z = k*dz
                ux[i,j,k] = 2+ 2*np.sin(z) + 8*np.sin(x)*np.cos(y)*np.cos(z)

N = (8,8,8)
kmax = int(np.ceil(np.linalg.norm(N))//2)+2
ek = np.zeros(kmax)

dx = 2*np.pi/N[0]
dy = 2*np.pi/N[1]
dz = 2*np.pi/N[2]
ux = cp.zeros(N)
set_vel_field(N, dx, dy, dz, ux)

ukx = np.zeros(N, dtype=np.complex64)
ukx = np.fft.rfftn(ux)/(N[0]*N[1]*N[2]) 

import math
BPG_x = math.ceil(N[0]/(TPB[0]))
BPG_y = math.ceil(N[1]/(TPB[1]))
BPG_z = math.ceil(N[2]/(TPB[2]))

BPG = (BPG_x, BPG_y, BPG_z)
print(BPG)


#Copy the arrays to the device
ukx_global_mem = cuda.to_device(ukx)
ek_global_mem = cuda.to_device(ek)

#start the kernel
compute_spectrum[BPG,TPB](N, ukx_global_mem, ek_global_mem)
cuda.synchronize()

# Copy the result back to the host
ek = ek_global_mem.copy_to_host()

print(ek)
```
